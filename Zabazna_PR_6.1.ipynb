```python
from vega_datasets import data
```

```python
source = data.wheat()
source
```

## Завдання 1

```python
chart = alt.Chart(source).mark_bar().encode(
    x='year:N',
    y='wheat:Q',
    color=alt.condition(alt.datum.year == 1810, alt.value('orange'), alt.value('blue')),
).properties(
    title='Стовпчаста діаграма кількості пшениці за роками'
)

chart
```

### Діаграма збережена окремим файлом 

## Завдання 2

```python
chart = alt.Chart(source).mark_bar().encode(
    x='year:N',
    y='wheat:Q',
    color=alt.value('steelblue'),
).properties(
    width=600
)
chart
```

```python
rule = alt.Chart(source).mark_rule(color='red').encode(
    y='average(wheat):Q',
    size=alt.value(2)
)
rule
```

```python
final_chart = (chart+rule).properties(width=600, title='Виробництво пшениці за роками')
final_chart
```

```python
final_chart.save('final.chart.html')
```

## Завдання 3

```python
source = data.stocks()
source
```

```python
line_chart = alt.Chart(source).mark_line(point=True).encode(
    x='date:T',
    y='price:Q',
    color='symbol:N',  # Визначаємо кожну лінію за допомогою кольорів
    tooltip=['symbol:N', 'price:Q', 'date:T']  # Додаємо точки зі значеннями та датою на наведення миші
).properties(
    width=800,
    height=400
).interactive() 

line_chart
```

```python
line_chart = line_chart.encode(alt.Color('symbol:N', legend=alt.Legend(title='Компанія')))
line_chart
```

```python
line_chart.save('interactive_line_chart.html')
```
