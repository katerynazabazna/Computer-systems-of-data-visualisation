```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

### Перейменування стовпців

```python
new_column_names = {col: col.lower().replace(' ', '_') for col in data.columns}

# Перейменувати стовпці за допомогою методу rename та нових назв
data = data.rename(columns=new_column_names)
data
```

## Завдання 1

```python
sns.set_theme()

plt.figure(figsize=(10, 6))  
sns.histplot(data['age'], bins=30, kde=True) 
plt.xlabel('Age')  
plt.ylabel('Count')  
plt.title('Histogram of Age')  
plt.show()
```

## Завдання 2

```python
plt.figure(figsize=(12, 6))  
sns.countplot(x='item_purchased', data=data, order=data['item_purchased'].value_counts().index)  
plt.xlabel('item_purchased')  
plt.ylabel('Count') 
plt.xticks(rotation=45)  
plt.title('Distribution of Purchased Items') 
plt.show()
```

### Гістограма розподілу оцінок товарів

```python
color = "#00BFFF"
plt.figure(figsize=(8, 6))
sns.histplot(data['review_rating'], bins=10, kde=True, color=color)  
plt.xlabel('Review Rating') 
plt.ylabel('Count') 
plt.title('Distribution of Review Ratings')  
plt.show()
```

### Ядерна оцінка щільності розподілу оцінок товарів

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data['review_rating'], shade=True)
plt.xlabel('Review Rating') 
plt.ylabel('Density')
plt.title('Kernel Density Estimate of Review Ratings')
plt.show()
```

# Завдання 2

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Побудова графіка поверхні
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis')


ax.set_title('3D Surface Plot of z=sin(sqrt(x^2+y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10, label='Z')

plt.show()
```
