```python
import pandas as pd
import numpy as np
import altair as alt
```

## Завдання 3.1

```python
x = np.arange(0, 100, 1)
y = x * np.cos(x)
data = pd.DataFrame({'x': x, 'y': y})
```

```python
line_chart = alt.Chart(data).mark_line().encode(
    x='x:Q',
    y='y:Q'   
).properties(
    title='Графік функції y = x * cos(x)',
    width=500,
    height=300 
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
)
line_chart
```

```python
line_chart = line_chart.mark_text(
    align='left',
    baseline='bottom',
    dx=5, 
    dy=-5,  
).encode(
    text='x:N'  
)
line_chart
```

## Завдання 3.3

```python
employment_data = data.us_employment()
employment_data
```

```python
histogram = alt.Chart(employment_data).mark_bar().encode(
    x='month:T',  
    y='nonfarm_change:Q',  
    color=alt.condition(alt.datum.nonfarm_change > 0, alt.value('green'), alt.value('red')),  # Колір для від'ємних та додатніх значень
).properties(
    title='Зміни в зайнятості в США',
    width=500  
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=16
)

histogram
```

## Завдання 3.4

```python
barley_data = data.barley()

histogram = alt.Chart(barley_data).mark_bar().encode(
    x='year:N',  
    y='sum(yield):Q', 
    color='year:N', 
    column='site:N'  
).properties(
    title='Врожай ячменю з групуванням за роками та місцями',
    width=alt.Step(40)  
).configure_axis(
    labelFontSize=12,
    titleFontSize=14
).configure_title(
    fontSize=16
)

histogram
```

## Завдання 3.7

```python
movies_data = data.movies.url

scatterplot = alt.Chart(movies_data).mark_point().encode(
    x=alt.X('IMDB_Rating:Q', bin=alt.Bin(maxbins=30), title='IMDB Rating'),  # Вісь X
    y=alt.Y('Rotten_Tomatoes_Rating:Q', bin=alt.Bin(maxbins=30), title='Rotten Tomatoes Rating'),  # Вісь Y
    color=alt.Color('count()', scale=alt.Scale(scheme='greenblue'), title='Count')  # Колір відображає кількість фільмів
).properties(
    width=600,
    height=400,
    title='Двовимірна гістограма рейтингів фільмів'
)

scatterplot
```

## Завдання 4

```python
file_path = "Global_Education.csv"
data = pd.read_csv(file_path, encoding='latin1')

print("Розмір датасету:", data.shape)
print("\nТипи даних:")
print(data.dtypes)

print("\nПропуски:")
print(data.isnull().sum())

print("\nДублікати:", data.duplicated().sum())

print("\nНазви стовпців:")
print(data.columns)
```

## Завдання 4.1 Описовий аналіз

```python
print("Перші 5 рядків:")
print(data.head())

print("\nОстанні 5 рядків:")
print(data.tail())

print("\nОсновна статистична інформація:")
print(data.describe())
```

## Завдання 4.2 Робота з числовими даними

```python
# Середнє значення та стандартне відхилення для числових стовпців
numeric_columns = ['Longitude', 'Birth_Rate']
numeric_stats_selected = data[numeric_columns].describe()

print("Статистика для вибраних числових стовпців:")
print(numeric_stats_selected)
```

```python
# Країна з найвищим та найнижчим рівнем безробіття
max_unemployment_country = data.loc[data['Unemployment_Rate'].idxmax(), 'Countries and areas']
min_unemployment_country = data.loc[data['Unemployment_Rate'].idxmin(), 'Countries and areas']

print("\nКраїна з найвищим рівнем безробіття:", max_unemployment_country)
print("Країна з найнижчим рівнем безробіття:", min_unemployment_country)
```

## Завдання 4.3 Робота з категоріальними даними

```python
unique_countries = data['Countries and areas'].unique()
print("Unique countries:", unique_countries)
```

```python
num_countries = len(unique_countries)
print("Number of countries:", num_countries)
```

## Завдання 4.4 Візуалізація даних

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['Gross_Primary_Education_Enrollment'], data['Gross_Tertiary_Education_Enrollment'], alpha=0.5)
plt.title('Scatter Plot: Gross Primary vs Tertiary Education Enrollment')
plt.xlabel('Gross Primary Education Enrollment')
plt.ylabel('Gross Tertiary Education Enrollment')
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 40))
bar_width = 0.35
bar_positions = range(len(data))

ax.barh(bar_positions, data['Youth_15_24_Literacy_Rate_Male'], bar_width, label='Male Literacy Rate')
ax.barh([pos + bar_width for pos in bar_positions], data['Youth_15_24_Literacy_Rate_Female'], bar_width, label='Female Literacy Rate')

ax.set_yticks([pos + bar_width / 2 for pos in bar_positions])
ax.set_yticklabels(data['Countries and areas'])
ax.set_xlabel('Literacy Rate')
ax.set_title('Literacy Rates for Youth (15-24) - Male vs Female')
ax.legend()

plt.show()
```

## Завдання 4.5 Маніпуляції з даними

```python
data['Total_Literacy_Rate'] = (data['Youth_15_24_Literacy_Rate_Male'] + data['Youth_15_24_Literacy_Rate_Female']) / 2

highest_literacy_country = data.loc[data['Total_Literacy_Rate'].idxmax(), 'Countries and areas']

print(f"The country with the highest overall literacy rate is: {highest_literacy_country}")
```
