### Завдання 9

```python
import calendar

# Припустимо, що у вас є DataFrame data з колонками 'date', 'avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center'

# Вибираємо дані для 2015 року
data_2015 = data[data['date'].dt.year == 2015]

# Групуємо дані за місяцями та обчислюємо суму відвідувачів кожного музею для кожного місяця
monthly_visits_2015 = data_2015.groupby(data_2015['date'].dt.month).sum()[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']]

# Створюємо словник для відображення чисел місяців у назви
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}

# Знаходимо місяць з найвищою загальною кількістю відвідувачів
highest_month = month_names[monthly_visits_2015.sum(axis=1).idxmax()]

# Знаходимо місяць з найнижчою загальною кількістю відвідувачів
lowest_month = month_names[monthly_visits_2015.sum(axis=1).idxmin()]

print('Місяць з найвищою загальною кількістю відвідувачів у 2015 році:', highest_month)
print('Місяць з найнижчою загальною кількістю відвідувачів у 2015 році:', lowest_month)
```

### Завдання 10

```python
summer_months = [6, 7, 8]  
winter_months = [12, 1, 2]

# Вибір даних для 2018 року
data2018 = data[data['date'].dt.year == 2018]

# Фільтрація за літо та зиму для 2018 року
summer_data = data2018[data2018['date'].dt.month.isin(summer_months)]
winter_data = data2018[data2018['date'].dt.month.isin(winter_months)]

# Знаходження середньої кількості відвідувачів для кожного періоду
summer_visitors = summer_data['avila_adobe'].sum()
winter_visitors = winter_data['avila_adobe'].sum()

print('Кількість відвідувачів музею Avila Adobe у літні місяці 2018 року:', summer_visitors, '\n')
print('Кількість відвідувачів музею Avila Adobe у зимові місяці 2018 року:', winter_visitors)
```

### Завдання 11

```python
data2016 = data[data['date'].dt.year == 2016]

# Вибір стовпців з кількістю відвідувачів кожного музею та датою
museum_visitors_columns = ['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']
museum_data_2016 = data2016[museum_visitors_columns + ['date']]

# Обчислення кореляцій
correlation_matrix = museum_data_2016.corr()

print('Кореляційна матриця:')
print(correlation_matrix)
```

```python
data2014 = data[data['date'].dt.year == 2014]

# Групуємо дані за місяцями та обчислюємо суму відвідувачів кожного музею для кожного місяця
monthly_visits_2014 = data2014.groupby(data2014['date'].dt.month).sum()[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']]
monthly_visits_2015 = data2015.groupby(data2015['date'].dt.month).sum()[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']]
monthly_visits_2016 = data2016.groupby(data2016['date'].dt.month).sum()[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']]
monthly_visits_2017 = data2017.groupby(data2017['date'].dt.month).sum()[['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']]

# Створюємо графіки
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.flatten()

# Графік за 2014 рік
monthly_visits_2014.plot(kind='bar', ax=axes[0], color=['blue', 'green', 'yellow', 'gray'])
axes[0].set_title('2014')
axes[0].set_ylabel('Кількість відвідувачів')

# Графік за 2015 рік
monthly_visits_2015.plot(kind='bar', ax=axes[1], color=['blue', 'green', 'yellow', 'gray'])
axes[1].set_title('2015')
axes[1].set_ylabel('Кількість відвідувачів')

# Графік за 2016 рік
monthly_visits_2016.plot(kind='bar', ax=axes[2], color=['blue', 'green', 'yellow', 'gray'])
axes[2].set_title('2016')
axes[2].set_ylabel('Кількість відвідувачів')

# Графік за 2017 рік
monthly_visits_2017.plot(kind='bar', ax=axes[3], color=['blue', 'green', 'yellow', 'gray'])
axes[3].set_title('2017')
axes[3].set_ylabel('Кількість відвідувачів')

plt.tight_layout()
plt.show()
```
