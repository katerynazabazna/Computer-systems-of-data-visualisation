```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Завдання 1
data = pd.read_csv("Global YouTube Statistics.csv", encoding='latin1')
data
```

```python
# Завдання 2
data.head(5)
```

```python
# Завдання 3
data.shape
```

```python
# Завдання 4
data.isna().sum()
```

```python
# Завдання 5
# Так як в нашому датафреймі є NaN в стовпцях типу object, 
# то ми замінюємо NaN на середнє значення тільки в стовпцях числового типу

data.replace("—", np.nan, inplace=True)
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
data
```

```python
data[numeric_columns] = data[numeric_columns].astype(float)
data.dtypes
```

```python
# Завдання 6 
data.isna().sum()
# Тепер в датафреймі пропуски лише в стовпцях типу object
```

```python
# Завдання 7
unique_values = data['Country'].unique()
print(unique_values)
```

```python
# Завдання 8 
data['video views'].plot(kind='hist', bins=10, edgecolor='black')
plt.xlabel('Перегляди')
plt.ylabel('Частота')
plt.title('Розподіл переглядів')
plt.show()
```
