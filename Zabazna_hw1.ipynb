## Завдання 5

Замініть типи нечислових стовпців числовими стовпцями. Це стовпці, у яких було розташовано символ "—"

### Дані зчитались з самого початку з типом int або float. Розглянемо приклад, якби їх дійсно потрібно було б перетворювати 

```python
numeric_columns = [col for col in table.columns if col != 'регіон']
table[numeric_columns] = table[numeric_columns].apply(pd.to_numeric, errors='coerce')
table
```

## Завдання 9

Отримайте список регіонів, де народжуваність у 2014 році була вищою за середню по Україні

```python
##Розраховуємо середнє значення за 2014 рік
mean_value = table_["2014"].mean()
mean_value
```
